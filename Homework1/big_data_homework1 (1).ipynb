```python
variables = ['income', 'gender', 'gamer']

# Store results
results = []
```

```python
print(f"{'Variable':<10} {'Test Mean':>10} {'Control Mean':>15} {'% Diff':>10} {'P-Value':>10}")
print("-" * 55)

# Loop through variables
for var in variables:
    test_mean = test_group[var].mean()
    control_mean = control_group[var].mean()
    percent_diff = 100 * (test_mean - control_mean) / control_mean
    t_stat, p_val = ttest_ind(test_group[var], control_group[var], equal_var=False)

    print(f"{var:<10} {test_mean:10.4f} {control_mean:15.4f} {percent_diff:10.2f}% {p_val:10.4f}")
```

```python
outcome_rate_winners = winners['outcome'].mean()
outcome_rate_losers = losers['outcome'].mean()

# Enrollment rate among lottery winners
enrollment_rate_winners = winners['treatment'].mean()
wald_estimate = (outcome_rate_winners - outcome_rate_losers) / enrollment_rate_winners

{
    "Outcome Rate (Winners)": round(outcome_rate_winners, 4),
    "Outcome Rate (Losers)": round(outcome_rate_losers, 4),
    "Enrollment Rate (Winners)": round(enrollment_rate_winners, 4),
    "Wald Estimate": round(wald_estimate, 4)
}
```

The walder estimate is -000.32

To treat the Wald estimate as a valid causal effect of Medicaid, a few key conditions must be true:

Winning the lottery should influence enrollment — this is true here because about 80% of winners actually enrolled.

Winning the lottery should only affect health outcomes through Medicaid enrollment — this means it shouldn’t change health outcomes in any other way (like making people more health-conscious just by winning). This is a strong assumption and may not always hold.

The lottery must be truly random — this is true in our case since it was a real lottery system.

There should be no one who would have enrolled only if they lost but not if they won — this is a technical condition called monotonicity and seems reasonable in this setting.

Q4c: Compare Standard Errors

```python
n_winners = len(winners)
n_losers = len(losers)

itt = outcome_rate_winners - outcome_rate_losers

se_itt = np.sqrt(
    (outcome_rate_winners * (1 - outcome_rate_winners)) / n_winners + (outcome_rate_losers * (1 - outcome_rate_losers)) / n_losers
)
```

```python
p_enroll = enrollment_rate_winners
```

```python
wald = itt / p_enroll
```

```python
se_wald = se_itt / p_enroll

{
    "ITT Estimate": round(itt, 4),
    "Standard Error (ITT)": round(se_itt, 5),
    "Wald Estimate": round(wald, 4),
    "Standard Error (Wald)": round(se_wald, 5)
}
```

The standard error for the Wald estimate (0.00164) is larger than the ITT standard error (0.00131).

This happens because we divide the ITT by the enrollment rate (which is less than 1), so the variance increases.

These standard errors could still be biased if assumptions like random sampling, correct model specification, or independence don’t fully hold in practice.
